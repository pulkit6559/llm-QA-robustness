```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data files
social_iqa_atomic_labels = pd.read_csv('Final_SocialIQA_with_ATOMIC_Labels.csv')
updated_y_m_o = pd.read_csv('updated_Y_M_O_par_similarity.csv')
updated_m_f_a = pd.read_csv('updated_M_F_A_par_similarity.csv')

# Add the 'Refined_ATOMIC_Label' to the updated datasets based on index alignment
updated_y_m_o['Refined_ATOMIC_Label'] = social_iqa_atomic_labels['Refined_ATOMIC_Label']
updated_m_f_a['Refined_ATOMIC_Label'] = social_iqa_atomic_labels['Refined_ATOMIC_Label']

# Function to calculate accuracy using aligned datasets
def calculate_accuracy_aligned(data, demographic_labels):
    accuracies = {}
    for label in demographic_labels:
        correct_predictions = data[data[f'choice_{label}'] == data['target']]
        accuracies[label] = correct_predictions.groupby('Refined_ATOMIC_Label').size() / data.groupby('Refined_ATOMIC_Label').size()
    return accuracies

# Calculate accuracies for each demographic set
demographic_labels_y_m_o = ['y', 'm', 'o']
demographic_labels_m_f_a = ['m', 'f', 'amb']

accuracy_y_m_o = calculate_accuracy_aligned(updated_y_m_o, demographic_labels_y_m_o)
accuracy_m_f_a = calculate_accuracy_aligned(updated_m_f_a, demographic_labels_m_f_a)

# Calculate baseline accuracy
baseline_accuracy = updated_y_m_o[updated_y_m_o['choice_b'] == updated_y_m_o['target']].groupby('Refined_ATOMIC_Label').size() / updated_y_m_o.groupby('Refined_ATOMIC_Label').size()

# Plotting function
def plot_final_adjusted_combined_accuracy_graph(accuracy_dict_y_m_o, accuracy_dict_m_f_a, baseline_acc, title):
    plt.figure(figsize=(20, 12))
    labels = ['oEffect', 'oWant', 'xAttr', 'xIntent', 'xNeed', 'xReact']
    
    colors = {
        'Young': '#1f77b4',  # blue
        'Middle': '#ff7f0e',  # orange
        'Old': '#2ca02c',  # green
        'Male': '#d62728',  # red
        'Female': '#9467bd',  # purple
        'Ambiguous': '#8c564b',  # brown
        'Baseline': '#343a40'  # dark grey
    }
    
    demographic_names_y_m_o = ['Young', 'Middle', 'Old']
    demographic_names_m_f_a = ['Male', 'Female', 'Ambiguous']
    
    for demographic, name in zip(demographic_labels_y_m_o, demographic_names_y_m_o):
        accuracies = accuracy_dict_y_m_o[demographic]
        plt.plot(labels, [accuracies.get(label, 0) for label in labels], label=name, color=colors[name])
    
    for demographic, name in zip(demographic_labels_m_f_a, demographic_names_m_f_a):
        accuracies = accuracy_dict_m_f_a[demographic]
        plt.plot(labels, [accuracies.get(label, 0) for label in labels], label=name, color=colors[name])
    
    plt.plot(labels, [baseline_acc.get(label, 0) for label in labels], label='Baseline', color=colors['Baseline'], linestyle='--')
    
    plt.title(title, fontsize=LARGE)
    plt.xlabel('ATOMIC Labels', fontsize=SMALL)
    plt.ylabel('Accuracy', fontsize=SMALL)
    plt.legend(fontsize=SMALL, title_fontsize=SMALL)
    plt.grid(True)
    plt.ylim(0.45, 0.7)
    plt.xticks(fontsize=SMALL)
    plt.yticks(fontsize=SMALL)
    
    age_plot_path = 'atomic_accuracies.png'
    plt.savefig(age_plot_path, dpi=300)
    
    plt.show()

# Plot the graph
plot_final_adjusted_combined_accuracy_graph(accuracy_y_m_o, accuracy_m_f_a, baseline_accuracy, "ATOMIC Accuracy for Demographic Sets")
```
